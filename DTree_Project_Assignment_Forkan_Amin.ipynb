** Do the Train Test Split** using test size 0.3

```python
# Features and target

X = df.drop('Target', axis=1)
y = df['Target']
```

```python
# Train-Test Split (30% test data)

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Check shapes

print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

### Decision Tree
**Importing decision tree classifier and create its instance**

```python
from sklearn.tree import DecisionTreeClassifier

# Create Decision Tree model instance
dt_model = DecisionTreeClassifier(random_state=42)
```

**Fit the training data to the model**

```python
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
```

**What is the Score of the model?**

```python
score = dt_model.score(X_test, y_test)
print("Model Accuracy:", score)
```

The Decision Tree classifier achieved an accuracy of 75.10% on the test dataset.

```python
from sklearn.metrics import confusion_matrix, classification_report

y_pred = dt_model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
